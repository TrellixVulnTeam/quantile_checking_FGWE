```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("~/Desktop/test.csv")
```

```python
y = np.array(df)[:,2][:1200] / 1e9
t = np.arange(0, len(y))
```

```python
size = (15,7.5)
```

```python
plt.figure(figsize=size, dpi=100)
plt.plot(t, y, label="Input")
plt.fill_between(t, 0, y, alpha=0.2)
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Input Bandwith$[Mbit / s]$')
plt.title(r'Grafico del InputBandwidth')
plt.show()
```

```python
_sorted = np.sort(y)
```

```python
plt.figure(figsize=size, dpi=100)
plt.plot(t, _sorted, label="Input")
plt.fill_between(t, 0, _sorted, alpha=0.2)
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Input Bandwith$[Mbit / s]$')
plt.title(r'Grafico del InputBandwidth Ordinato')
plt.show()
```

```python
q = np.quantile(y, 0.95)
print(q)
```

```python
i = np.searchsorted(_sorted, q)
print(i)
```

```python
n = 100
v = np.linspace(min(t) - 10, i, n)
z = [q] * n
```

```python
plt.figure(figsize=size, dpi=100)
plt.plot(t[:i], _sorted[:i])
plt.fill_between(t[:i], 0, _sorted[:i], alpha=0.2)
plt.plot(t[i:], _sorted[i:])
plt.fill_between(t[i:], 0, _sorted[i:], alpha=0.2)

plt.plot(v, z, "--")
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Input Bandwith$[Mbit / s]$')
plt.title(r'95-th percentile ~= 26')
plt.text(1170, 11, " 5% Del totale", weight = 'bold',rotation=90, horizontalalignment="center")
plt.text(1120, 11, "95% Del totale", weight = 'bold',rotation=90, horizontalalignment="center")
plt.show()
```

```python
sotto = y < q
sotto_y = y[sotto]
sotto_t = t[sotto]
sopra = y >= q
sopra_y = y[sopra]
sopra_t = t[sopra]
```

```python
plt.figure(figsize=size, dpi=100)
plt.plot(t, y)
plt.fill_between(sotto_t, 0, sotto_y, alpha=0.2)
plt.plot(sopra_t, sopra_y,  ".")
plt.plot([0,1300], [q,q], "--")
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Input Bandwith$[Mbit / s]$')
plt.title(r'Grafico del InputBandwidth')
plt.show()
```

# IN and OUT

```python
g = np.concatenate([y[:200]] * 6)
plt.figure(figsize=size, dpi=100)
plt.plot(t, y, label="Input")
plt.fill_between(t, 0, y, alpha=0.5)
plt.plot(t, g, label="Output")
plt.fill_between(t, 0, g, alpha=0.5)
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Bandwith$[Mbit / s]$')
plt.title(r'Grafico di Input ed Output Bandwidth')
plt.show()
```

```python
m = [max(x, k) for x,k in zip(g,y)]
plt.figure(figsize=size, dpi=100)
plt.plot(t, y, "--", label="Input")
plt.fill_between(t, 0, y, alpha=0.5)
plt.plot(t, g, "--", label="Output")
plt.fill_between(t, 0, g, alpha=0.5)
plt.plot(t, m, color="red", label="Max")
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Bandwith$[Mbit / s]$')
plt.title(r'Grafico di Input ed Output Bandwidth e Massimo Puntuale')
plt.show()
```

```python
m = [max(x, k) for x,k in zip(g,y)]
plt.figure(figsize=size, dpi=100)
plt.plot(t, m, color="red", label="Max")
plt.fill_between(t, 0, m, alpha=0.5, color="red")
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Max Bandwith$[Mbit / s]$')
plt.title(r'Grafico del Massimo Puntuale')
plt.show()
```

# Multipli

```python
y = y[:20]
t = t[:20]
```

```python
plt.figure(figsize=size, dpi=100)
plt.plot(t, y, ".-",label="Max servizio1")
plt.fill_between(t, 0, y, alpha=0.2)
plt.plot(t+0.4, 15-y, ".-", label="Max servizio2")
plt.fill_between(t+0.4, 0, 15-y, alpha=0.2)
plt.legend()
plt.xlim(0-0.5, len(y)+0.5)
plt.ylim(0, max(y)+0.5)
plt.xlabel('$t$')
plt.ylabel(r'Bandwith$[Mbit / s]$')
plt.title(r'Grafico di due servizi coi tempi non allineati')
plt.show()
```
